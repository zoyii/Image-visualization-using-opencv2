**Resize**

```python
resized_img = cv2.resize(img, (100, 100), interpolation=cv2.INTER_NEAREST)
plt.imshow(resized_img)
```

**Rotate**

```python
rotated_img = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(rotated_img)
```

**Translation**

```python
import numpy as np
```

```python
translation_arr = np.float32([[1, 0, -10], [0, 1, -10]])
```

```python
translated_img = cv2.warpAffine(img, translation_arr, (1000, 1000))
```

```python
plt.imshow(translated_img)
```

**Concatenation**

```python
!ls
```

```python
main_office = cv2.imread('main.jpg')
library = cv2.imread('library.jpg')
```

```python
main_office = cv2.resize(main_office, (100, 100))
library = cv2.resize(library, (100, 100))
```

```python
combined = cv2.hconcat([main_office, library])
```

```python
plt.imshow(combined)
```

**BGR TO RGB**

```python
real_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(real_img)
```

**GrayScale Conversion**

```python
gray_img = cv2.cvtColor(real_img, cv2.COLOR_RGB2GRAY)
```

```python
gray_img.shape  # now it has single channels
```

```python
real_img.shape  # real image has 3 channels
```

```python
plt.imshow(gray_img, cmap='gray')
```

**Edge Detection**

```python
edges = cv2.Canny(real_img, 100, 200)
```

```python
plt.imshow(edges, cmap='gray')
```

**Save Image**

```python
cv2.imwrite('artistic.jpg', edges)
```
